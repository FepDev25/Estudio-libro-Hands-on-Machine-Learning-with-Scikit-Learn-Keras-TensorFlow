# Exercises

In this chapter we have covered some of the most important concepts in machine learning. In the next chapters we will dive deeper and write more code, but before we do, make sure you can answer the following questions:

### 1. How would you define machine learning?

- Machine learning is the ability of computers to learn from data. Learning is performing a task correctly.

### 2. Can you name four types of applications where it shines?

- Language recognition, sales prediction, image classification, and a personal assistant.

### 3. What is a labeled training set?

- A dataset where each row is categorized, belongs to a specific category.

### 4. What are the two most common supervised tasks?

- Regression and classification.

### 5. Can you name four common unsupervised tasks?

- Clustering, visualization, dimensionality reduction, and association rule learning.

### 6. What type of algorithm would you use to allow a robot to walk in various unknown terrains?

- Reinforcement learning.

### 7. What type of algorithm would you use to segment your customers into multiple groups?

- Clustering algorithms if you do not have labeled data, or classification algorithms if you do.

### 8. Would you frame the problem of spam detection as a supervised learning problem or an unsupervised learning problem?

- Supervised learning problem.

### 9. What is an online learning system?

- It is a type of system where the model can be trained by adding more data once it has already been trained, making it an incremental training process.

### 10. What is out-of-core learning?

- Can handle vast quantities of data that cannot fit in a computer's main memory.

### 11. What type of algorithm relies on a similarity measure to make predictions?

- An instance-based learning

### 12. What is the difference between a model parameter and a model hyperparameter?

- You configure the hyperparameters before starting to control how the model learns, while the parameters are the values ​​that the model learns by itself during training.

### 13. What do model-based algorithms search for? What is the most common strategy they use to succeed? How do they make predictions?

- Model-based algorithms seek to find optimal values ​​for their internal parameters so that the model generalizes well to new data. Their main strategy for achieving this is to minimize a cost function during training (which measures error and sometimes penalizes complexity), and to make predictions, they simply incorporate the characteristics of the new instance into the model function using these already adjusted parameters.

### 14. Can you name four of the main challenges in machine learning?

- Insufficient Quantity of Training Data, Nonrepresentative Training Data, Poor-Quality Data and Irrelevant Features.

### 15. If your model performs great on the training data but generalizes poorly to new instances, what is happening? Can you name three possible solutions?

- Overfitting. Get more training data, reduce the model complexity, or increase the regularization.

### 16. What is a test set, and why would you want to use it?

- The sety test is used to measure the error of model predictions with new instances.

### 17. What is the purpose of a validation set?

- Compare models, to select the best one and adjust hyperparameters.

### 18. What is the train-dev set, when do you need it, and how do you use it?

- The train-dev set is used when there is a risk of discrepancy between the training data and the data used in the validation and test datasets. The train-dev set is a reserved portion of the training set. The model is trained on the rest of the training set and evaluated on both the train-dev and validation sets. If the model performs well on the training set but not on the train-dev set, you are likely overfitting the training set. If it performs well on both the training and train-dev sets but not on the validation set, there is probably a significant discrepancy between the training data and the validation and test data, and you should try to improve the training data to more closely resemble the validation and test data.

### 19. What can go wrong if you tune hyperparameters using the test set?

- Overfitting the test set.
